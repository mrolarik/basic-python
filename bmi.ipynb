# เขียนโปรแกรมภาษาไพธอนเพื่อคำนวณค่าดัชนีมวลกาย (Body Mass Index: BMI)

* รูปภาพจาก http://www.bangpakokhospital.com/care_blog/content/วีธีคำนวณดัชนีมวลกายBMI

จากตัวอย่างข้างต้น ค่า BMI สามารถกำหนดได้เป็น 5 ช่วง ดังนี้


*   **BMI > 30.0** คือ โรคอ้วนอันตราย (Dangerous obesity)
*   **BMI > 25.0** คือ โรคอ้วน (Obesity)
*   **BMI > 23.0** คือน้ำหนักเกิน (Over weight)
*   **BMI > 18.5** สมส่วน (Slim)
*   **BMI <= 18.5** น้ำหนักต่ำกว่าเกณฑ์ (lower weight)

ในการคำนวณค่า BMI สามารถทำได้โดย

```
BMI = weight / (high * high)
```

โดยที่ **weight** คือ น้ำหนักกิโลกรัม (67 kg.)  
**high** คือส่วนสูงคำนวณเป็นเมตร (175 cm. คือ 1.75 m.)



## การเขียนโปรแกรม BMI ทำได้ดังนี้

### สร้าง input เพื่อรับข้อมูล weight และ high

````
# input
# ใช้ฟังก์ชัน float() เพื่อเปลี่ยนค่าที่รับจาก input() ให้เป็นข้อมูลประเภท floating point 
# เก็บข้อมูลที่ป้อนจากคีย์บอร์ดไว้ที่ตัวแปร weight
weight = float(input('Input your weight (kg.) : '))
high = float(input('Input your high (m.) : '))
````

### คำนวณค่า BMI

````
# calcluation
BMI = weight / (high * high)

# output
print('Your BMI is ', BMI)
````

### สร้างเงื่อนไข if-else เพื่อแสดงผลลัพธ์

* BMI > 30.0 คือ โรคอ้วนอันตราย (Dangerous obesity)
* BMI > 25.0 คือ โรคอ้วน (Obesity)
* BMI > 23.0 คือน้ำหนักเกิน (Over weight)
* BMI > 18.5 สมส่วน (Slim)
* BMI <= 18.5 น้ำหนักต่ำกว่าเกณฑ์ (lower weight)

````
# นำค่า BMI มาเข้าเงื่อนไข if-else
if(BMI > 30.0):
  print('Dangerous obesity')
elif(BMI > 25.0):
  print('Obesity')
elif(BMI > 23.0):
  print('Over weight')
elif(BMI > 18.5):
  print('Slim')
else:
  print('lower weight')
````

### ใช้ while-loop เพื่อให้กรอกข้อมูลได้เรื่อย ๆ


````
while(True):
  weight = float(input('Input your weight (kg.) : '))
  high = float(input('Input your high (m.) : '))
  
  BMI = weight / (high * high)
  print('Your BMI is ', BMI)
  
  if(BMI > 30.0):
    print('Dangerous obesity')
  elif(BMI > 25.0):
    print('Obesity')
  elif(BMI > 23.0):
    print('Over weight')
  elif(BMI > 18.5):
    print('Slim')
  else:
    print('lower weight')
````

* จากกรณีข้างต้น ใช้ while-loop โดยการกำหนดให้มีค่าเป็นจริง (True)ในการวนรอบแบบไม่จบ (infinity loop) 

```
while(True):
```

* ดังนั้น หากต้องการให้ whle-loop หยุดการทำงานจะต้องสร้างเงื่อนไข เช่น

```
import sys

chk = input('Do you want to exit (y/n)? ')

if(chk == 'y'):
    sys.exit()
```

* จากตัวอย่าง กำหนดให้ป้อนข้อมูลทางคีย์บอร์ด และนำค่าที่ได้มาเก็บไว้ที่ตัวแปร **chk**
* จากนั้นนำตัวแปร **chk** ไปเข้าเงื่อนไข **if** หากค่าที่ป้อนเข้ามามีค่าเป็น **y** เงื่อนไขก็จะเป็นจริง (True) ดังนั้น โปรแกรมจะหยุดทำงานด้วยคำสั่ง ```sys.exit()```


````
import sys

while(True):
  weight = float(input('Input your weight (kg.) : '))
  high = float(input('Input your high (m.) : '))
  
  BMI = weight / (high * high)
  print('Your BMI is ', BMI)
  
  if(BMI > 30.0):
    print('Dangerous obesity')
  elif(BMI > 25.0):
    print('Obesity')
  elif(BMI > 23.0):
    print('Over weight')
  elif(BMI > 18.5):
    print('Slim')
  else:
    print('lower weight')
    
  chk = input('Do you want to exit (y/n)? ')

  if(chk == 'y'):
    sys.exit()
````

* จากกรณีข้างต้นการใช้คำสั่ง ```sys.exit()``` จะไม่เกิดข้อผิดพลาดหากทดสอบผ่าน Terminal หรือ cmd
* แต่กรณีนี้ทดสอบใน jupyter notebook จึงเกิดข้อผิดพลาด

### การเพิ่มข้อมูลลงไปใน list

* ในกรณีนี้ จะเพิ่มข้อมูล name, weight, high และ bmi ลงไปเก็บไว้ใน list
* ดังนั้นให้สร้าง list จำนวน 4 ตัวแปร ประกอบด้วย name_list, weight_list, high_list และ bmi_list

````
import sys

name_list = []
weight_list = []
high_list = []
bmi_list = []

while(True):
  name = input('Input your name : ')
  weight = float(input('Input your weight (kg.) : '))
  high = float(input('Input your high (m.) : '))
  
  BMI = weight / (high * high)
  
  #add to list
  name_list.append(name)
  weight_list.append(weight)
  high_list.append(high)
  bmi_list.append(BMI)
  
  print('Your BMI is ', BMI)
  
  if(BMI > 30.0):
    print('Dangerous obesity')
  elif(BMI > 25.0):
    print('Obesity')
  elif(BMI > 23.0):
    print('Over weight')
  elif(BMI > 18.5):
    print('Slim')
  else:
    print('lower weight')
    
  chk = input('Do you want to exit (y/n)? ')

  if(chk == 'y'):
    print('Name \t weight \t high \t BMI')
    for idx in range(len(name_list)):
      print(name_list[idx], '\t', weight_list[idx], '\t', high_list[idx], '\t', bmi_list[idx])
    
    sys.exit()
````

* จากตัวอย่างข้างต้น ต้องการแสดงข้อมูลทั้งหมดที่ป้อนลงไป โดยนำโค้ดไปวางไปในเงื่อนไข ``` if(chk == 'y'):```
